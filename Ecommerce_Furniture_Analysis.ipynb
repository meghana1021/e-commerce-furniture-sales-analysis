```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)

print("Libraries Imported Successfully")
```

## 📂 Load Dataset

```python
df = pd.read_csv('../dataset/e-commerce_furniture_analysis.csv')

df.head()
```

## 📊 Dataset Overview

```python
print("Rows and Columns:", df.shape)

print("\nColumns in Dataset:\n")
print(df.columns)

print("\nDataset Information:\n")
df.info()
```

## 🔍 Checking Missing Values

```python
missing_values = df.isnull().sum()

missing_values
```

## 🧹 Data Cleaning

```python
df['price'] = df['price'].astype(str).str.replace('$', '', regex=False)
df['price'] = pd.to_numeric(df['price'], errors='coerce')

df['originalPrice'] = df['originalPrice'].astype(str).str.replace('$', '', regex=False)
df['originalPrice'] = pd.to_numeric(df['originalPrice'], errors='coerce')

df['originalPrice'] = df['originalPrice'].fillna(df['price'])

df['tagText'] = df['tagText'].fillna(df['tagText'].mode()[0])

print("Missing Values After Cleaning:\n")
print(df.isnull().sum())
```

## 📈 Statistical Summary

```python
df.describe()
```

## 📊 Exploratory Data Analysis (EDA)

### 1️⃣ Top 10 Best Selling Products

```python
top_products = df.sort_values(by='sold', ascending=False).head(10)

plt.figure(figsize=(12,6))
sns.barplot(x='sold', y='productTitle', data=top_products)

plt.title('Top 10 Best Selling Products')
plt.xlabel('Units Sold')
plt.ylabel('Product Title')

plt.show()
```

### 2️⃣ Price Distribution

```python
plt.figure(figsize=(10,6))

plt.hist(df['price'], bins=30)

plt.title('Price Distribution')
plt.xlabel('Price')
plt.ylabel('Frequency')

plt.show()
```

### 3️⃣ Original Price vs Discounted Price

```python
plt.figure(figsize=(10,6))

plt.scatter(df['originalPrice'], df['price'])

plt.title('Original Price vs Discounted Price')
plt.xlabel('Original Price')
plt.ylabel('Discounted Price')

plt.show()
```

### 4️⃣ Sales Distribution

```python
plt.figure(figsize=(10,6))

plt.hist(df['sold'], bins=30)

plt.title('Sales Distribution')
plt.xlabel('Units Sold')
plt.ylabel('Frequency')

plt.show()
```

## 💡 Business Insights

### Key Insights
- Most furniture products are priced in the low-to-mid price range.
- Discounted products tend to achieve higher sales.
- Few products generate very high sales compared to others.
- Pricing strategy plays an important role in customer purchases.
- Best-selling products contribute significantly to overall sales.
